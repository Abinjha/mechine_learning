```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('face3.jpg')
```

```python
plt.imshow(img)
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
face=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
face
```

```python
faces=face.detectMultiScale(conv_img,minNeighbors=9)
faces
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(conv_img,(x,y),(x+w,y+h),(255,0,0),3)
plt.imshow(conv_img)
```
