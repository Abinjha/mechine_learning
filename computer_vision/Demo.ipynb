```python
#convert colour image to gray scale
img1_gray=cv2.imread('baby1.jpg',cv2.IMREAD_GRAYSCALE)
img1_gray
```

```python
cv2.imshow("gray_scale",img1_gray)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

```python
img2_gray=cv2.imread("baby2.jpg",cv2.IMREAD_GRAYSCALE)
cv2.imshow("gray_scale",img2_gray)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

```python
#save
cv2.imwrite('gray_scale.jpg',img1_gray)
```

```python
#convert into binary image(black & white)
#do not convert colour image directly into binary image
#first change into gary scale then change binary
#threshold======> it return two values

thresh,bin_img=cv2.threshold(img1_gray,180,255,cv2.THRESH_BINARY)   #thresh=180,255   bin_img=binary image
                             #0-180 pixel convert into black remaining pixel convert into white (maximum pixel is 255 it constant)                                                             
cv2.imshow("binary_img",bin_img)
cv2.waitKey(10000)
cv2.destroyAllWindows()



```

```python
img4=cv2.imread("baby1.jpg")

#display image
import matplotlib.pyplot as plt
plt.imshow(img4)
```

```python
#cv2=====>read===>RGB====>BGR
#show=====>cv2===>bgr===>rgb

#matplotlib=====>bgr
```

```python
#conver bgr===>rgb
conv_img=cv2.cvtColor(img4,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
img5=cv2.imread('baby2.jpg')
plt.imshow(img5)
```

```python
conv_img3=cv2.cvtColor(img5,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img3)
```

```python
#boxing plot
#rectangle
cv2.rectangle(conv_img3,pt1=(250,90),pt2=(710,610),color=(255,0,0),thickness=2)
plt.imshow(conv_img3)
```

```python
conv_img=cv2.cvtColor(img4,cv2.COLOR_BGR2RGB)
#circle
cv2.circle(conv_img,center=(360,320),radius=300,color=(0,255,0),thickness=3)
plt.imshow(conv_img)
```

```python
conv_img=cv2.cvtColor(img4,cv2.COLOR_BGR2RGB)
cv2.circle(conv_img,center=(360,320),radius=300,color=(0,255,0),thickness=-1)
plt.imshow(conv_img)
```

```python
img6=cv2.imread("baby3.jpg")
conv_img4=cv2.cvtColor(img6,cv2.COLOR_BGR2RGB)
#add text in to a image
cv2.putText(conv_img4,text='CUTE BABY',org=(200,300),fontFace=cv2.FONT_HERSHEY_COMPLEX,fontScale=10,color=(0,0,255),thickness=10)
plt.imshow(conv_img4)
```
