```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['Gender']=lab.fit_transform(df['Gender'])
df['Hx Radiothreapy']=lab.fit_transform(df['Hx Radiothreapy'])
df['Adenopathy']=lab.fit_transform(df['Adenopathy'])
df['Pathology']=lab.fit_transform(df['Pathology'])
df['Focality']=lab.fit_transform(df['Focality'])
df['Risk']=lab.fit_transform(df['Risk'])
df['T']=lab.fit_transform(df['T'])
df['N']=lab.fit_transform(df['N'])
df['M']=lab.fit_transform(df['M'])
df['Stage']=lab.fit_transform(df['Stage'])
df['Response']=lab.fit_transform(df['Response'])
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
scale.fit(x_train)
x_train=scale.transform(x_train)
x_test=scale.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
knn=KNeighborsClassifier(n_neighbors=7)
navi=BernoulliNB()
support=SVC()
lst=[knn,navi,support]
```

```python
for i in lst:
    i.fit(x_train,y_train)
    y_pred=i.predict(x_test)
    print("confusion metrix is.......")
    print(confusion_matrix(y_test,y_pred))
    print("accuracy score is....")
    print(accuracy_score(y_test,y_pred))
    print("classification report....")
    print(classification_report(y_test,y_pred))
    print('*'*100)
```
