```python
obj=df[['Make','Model','Vehicle Class','Transmission','Fuel Type']]
Y=df['CO2 Emissions(g/km)']

```

```python
from sklearn.preprocessing import LabelEncoder
for col in obj.columns:
    lab=LabelEncoder()
    obj[col]=lab.fit_transform(obj[col])
```

```python
from sklearn.feature_selection import chi2,SelectKBest
# chi_score=chi2(obj,y)
# chi_score
# p_values=pd.Series(chi_score[0])
# p_values.index=obj.columns
# p_values
```

```python
chi2_selector=SelectKBest(score_func=chi2,k='all')
chi2_selector.fit(obj,Y)
chi_score=chi2_selector.scores_
for feature,score in zip(obj.columns,chi_score):
    print(f"{feature}:{score:.2f}")
```

```python
df.drop(['Fuel Consumption Comb (mpg)','Fuel Type'],axis=1,inplace=True)
```

```python
import matplotlib.pyplot as plt
def removal_box_plot(df,column,threshold):    #argument
    sns.boxplot(df[column])    #df['total_bill]  #sns.boxplot(df['total_bill])
    plt.title(f'orginal_box_plot of {column}')  #40
    plt.show()
    
    removed_outliers=df[df[column]<=threshold]  

    sns.boxplot(removed_outliers[column])
    plt.title('Box plot with out outliers of {column}')
    plt.show
    return removed_outliers

threshold_value=450

no_outliers=removal_box_plot(df,'CO2 Emissions(g/km)',threshold_value)
```

```python
dfe1=pd.get_dummies(df[['Make','Model','Vehicle Class','Transmission']],drop_first=True,dtype=int)
df1=pd.concat([df,dfe1],axis=1)
df1.shape
```

```python
df1.drop(['Model','Vehicle Class','Make','Transmission'],axis=1,inplace=True)
```

```python
x=df1.drop(['CO2 Emissions(g/km)'],axis=1)
y=df1['CO2 Emissions(g/km)']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df2=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred,'difference is':y_test-y_pred})
df2
```

```python
from sklearn.metrics import  mean_absolute_error,mean_squared_error, root_mean_squared_error, mean_absolute_percentage_error,r2_score
print('MAE is',mean_absolute_error(y_test,y_pred))
print('MSE is',mean_squared_error(y_test,y_pred))
print('RMSE id',root_mean_squared_error(y_test,y_pred))
print('MAPE is',mean_absolute_percentage_error(y_test,y_pred))
print('R2score is',r2_score(y_test,y_pred))
```
