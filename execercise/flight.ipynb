```python
#get dummies
dfg=pd.get_dummies(df[['Airline','Source','Destination']],drop_first=True,dtype=int)
      
```

```python
df2=pd.concat([df,dfg],axis=1)
df2
```

```python
df2.drop(['Airline','Source','Destination'],axis=1,inplace=True)
df2.shape
```

```python
x=df2.drop(['Price'],axis=1)
y=df2['Price']
```

```python
from sklearn.model_selection import train_test_split
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df1=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred,'difference is':y_test-y_pred})
df1
```

```python
from sklearn.metrics import  mean_absolute_error,mean_squared_error, root_mean_squared_error, mean_absolute_percentage_error,r2_score
print('MAE is',mean_absolute_error(y_test,y_pred))
print('MSE is',mean_squared_error(y_test,y_pred))
print('RMSE id',root_mean_squared_error(y_test,y_pred))
print('MAPE is',mean_absolute_percentage_error(y_test,y_pred))
print('R2score is',r2_score(y_test,y_pred))
```
