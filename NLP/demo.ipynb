```python
sen='''The Natural Language Toolkit (NLTK) is an open source Python library
    for Natural Language Processing.  A free online book is available.'''
```

```python

from nltk.corpus import stopwords
stop=stopwords.words('english')
from nltk.tokenize import word_tokenize
word1=word_tokenize(sen)
stop
```

```python
#n-gram  (tokenization method)

from nltk.util import ngrams
sample='this is a very good book to study'
data=ngrams(sequence=nltk.word_tokenize(sample),n=2)
for i in data:
    print(i)
```

```python
#stemming
from nltk.stem import PorterStemmer
ps=PorterStemmer()
word2=['running','cycling','thought','worked','reached','looping']
for i in word2:
    print(i,":",ps.stem(i))
```

```python
#snowball stemer
from nltk.stem import SnowballStemmer
sb=SnowballStemmer('english')
for i in word2:
    print(i,":",sb.stem(i))
```

```python
#lematization
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
lematizer=WordNetLemmatizer()
print('rockes:',lematizer.lemmatize('rocks'))
print('running:',lematizer.lemmatize('running'))
print('colours:',lematizer.lemmatize('colours'))
print('reached',lematizer.lemmatize('reached'))
```
