```python
import numpy as np
a=np.array([4,6,8,9,10])
b=np.array([6,4,8,3,2])
plt.plot(a,b)
```

```python
c=np.array([7,5,6,9,4]) #====>mechine read it as y coordinate,
plt.plot(c)
```

```python
plt.plot(c,'*') #plot only points  markers===*,s,o,+
```

```python
#line with points
plt.plot(c,marker='*')
```

```python
#point,line and colour
plt.plot(c,'*:g')
#g=colours , : line style
```

```python
#marker properties
plt.plot(c,marker='o',ms=15,mfc='r',mec='g') 
#ms= marker size , mfc=marker face colour , mec=marker edge colour
```

```python
#line property
plt.plot(c,linewidth=7,linestyle='dashed',color='r',marker='o',ms=25,mfc='c')
```

```python
x=np.array([9,5,4,3])
plt.plot(x,linewidth=4,linestyle='dotted',color='b',marker='s',mfc='g',ms=15,mec='r')
```

```python
#more than one line in a graph
x1=[1,2,3,4]
x2=[3,4,2,6]
y1=[6,4,8,1]
y2=[6,8,3,1]
plt.plot(x1,y1)
plt.plot(x2,y2)
```

```python
#add labels in graph
hight=[150,151,154,156]
weight=[44,43,45,41]
plt.plot(hight,weight)
plt.xlabel("height")
plt.ylabel("weight")
plt.title("Height-Weight Graph")
```

```python
velocity=[24,25,34,54,65]
displacement=[4,6,8,12,11]
plt.plot(velocity,displacement,'co--',mfc='m')
plt.xlabel("Velocity",color='r',size=11)
plt.ylabel('Displacement',color='r',size=11)
plt.title('Velocity-Displacement graph',color='k',size=15,family='serif',weight='bold') #family/fontname
plt.grid(linestyle='dashed',color='r')
```

```python
#subplot
x1=[1,2,3,4]
y1=[3,7,1,9]
plt.subplot(1,2,1) # 3rd(2nd 1) represent  position   #(1,2) two column in one row
plt.plot(x1,y1,marker='*')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("graph1")

x2=[1,7,3,4]
y2=[3,8,1,2]
plt.subplot(1,2,2)
plt.plot(x2,y2,marker='o')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("graph2")

plt.suptitle('GRAPH')
```

```python
#scaterplot
x1=[5,6,8,9,3,2]
y1=[3,4,5,7,1,8]
plt.scatter(x1,y1)

x2=[1,2,3,4,5,6]
y2=[3,4,5,7,1,8]
plt.scatter(x2,y2)
```

```python
#barchart
x=[1,2,3,4,5,6]
y=[4,5,3,6,8,6]
plt.bar(x,y)
```

```python
country=['india','uk','aus','pak','nz','eng','wi']
trop_count=[10,4,6,8,3,2,8]
plt.barh(country,trop_count,color='m')  #barh horizontal
plt.xlabel('country_name')
plt.ylabel('trop_count')
plt.title('GRAPH')
```

```python
#pi chart
data1=['apple','orange','grape','mango']
data=[10,20,30,40]
plt.pie(data,labels=data1,autopct='%.1f%%') #autopct using for wright percentage in graph
plt.legend(loc='lower left')  #loc=upper left,lower left,upper right,lower right
plt.tight_layout() #using for avoid legend overlapping
```
