```python
import numpy as np
import pandas as pd
df=pd.read_csv('customer5.txt',header=None)
df.columns=['id','fname','lname','age','prof','location','salary']
df
```

```python
#each country count
x=df['location'].value_counts()
x
```

```python

import matplotlib.pyplot as plt
plt.pie(x,labels=x.index,autopct='%.1f%%')   #x.values=x
plt.legend(loc='upper left')  #loc=upper left,lower left,upper right,lower right
plt.tight_layout()
plt.title('country count',color='g')
```

```python
plt.bar(x.index,x,color='g')  #plt.bar(x.index,x.values)
plt.xlabel('country_name')
plt.ylabel('country_count')
plt.title('country')
x
```
