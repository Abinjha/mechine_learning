```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('poly.csv')
df
```

```python
df.dtypes
```

```python
x=df.iloc[:,1:2]
y=df.iloc[:,-1]
```

```python
a=df['Pressure'].value_counts
plt.bar
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
y_pred=model.predict(x)
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred)
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
print('r2 score is',r2_score(y,y_pred))
print('mean_absolute_percentage_error',mean_absolute_percentage_error(y,y_pred))
```

```python
#polynomial regression
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=3)
x_poly=poly.fit_transform(x)
x_poly
```

```python
model1=LinearRegression()
model1.fit(x_poly,y)
y_poly=model1.predict(x_poly)
y_poly
```

```python
df1=pd.DataFrame({'Actual value':y,'simple lin':y_pred,'poly':y_poly})
```

```python
plt.scatter(x,y)
plt.plot(x,y_poly)
```
