```python
import numpy as np
import pandas as pd
df=pd.read_csv('Student_Performance.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['Extracurricular Activities']=lab.fit_transform(df['Extracurricular Activities'])
df.dtypes
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
import seaborn as sns
sns.regplot(x=df['Hours Studied'],y=y)
```

```python

sns.regplot(x=df['Sleep Hours'],y=y)
```

```python

sns.regplot(x=df['Sample Question Papers Practiced'],y=y)
```

```python

sns.regplot(x=df['Previous Scores'],y=y,color='g')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df1=pd.DataFrame({'Actual_value':y_test,'predicted_value':y_pred})
df1
```

```python
from sklearn.metrics import mean_absolute_error
print('mean absolue error is',mean_absolute_error(y_test,y_pred))
```
